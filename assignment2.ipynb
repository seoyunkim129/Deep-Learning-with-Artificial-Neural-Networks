```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split

# 데이터 로드
train_data = np.load("/content/drive/MyDrive/인신딥/train.npz")
test_data = np.load("/content/drive/MyDrive/인신딥/test.npz")

x_train = train_data['x']
y_train = train_data['y']
x_test = test_data['x']

# 데이터 전처리
x_train = x_train / 255.0  # 이미지 픽셀 값을 0과 1 사이로 정규화
x_test = x_test / 255.0

# 데이터 차원 확장
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

# 데이터 증강 생성기
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# 데이터 분리
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# 더 복잡한 CNN 모델 구성
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(5, activation='softmax')
])

# 모델 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 데이터 증강을 적용하여 모델 훈련
history = model.fit(datagen.flow(x_train, y_train, batch_size=64), epochs=25, validation_data=(x_val, y_val))

# 시험 데이터 예측
predictions = model.predict(x_test)

# 결과를 파일에 저장
with open("assignment.txt", "w") as file:
    for label in np.argmax(predictions, axis=1):
        file.write(str(label) + "\n")

```
